```python
# Find the Longest Substring Containing Vowels in Even Counts
class Solution(object):
    def findTheLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        # Mapping vowels to bits
        vowels_map = {'a': 0, 'e': 1, 'i': 2, 'o': 3, 'u': 4}
        bitmask = 0
        max_length = 0
        
        # Dictionary to store the first occurrence of each bitmask
        seen = {0: -1}  # We have seen the bitmask 0 at index -1
        
        for i, char in enumerate(s):
            if char in vowels_map:
                # Toggle the corresponding bit for the vowel
                bitmask ^= (1 << vowels_map[char])
            
            if bitmask in seen:
                # If we've seen this bitmask before, calculate the length of the substring
                max_length = max(max_length, i - seen[bitmask])
            else:
                # Otherwise, store the first occurrence of this bitmask
                seen[bitmask] = i
        
        return max_length
    
# findTheLongestSubstring
class Solution(object):
    def findTheLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        vowels_List = set(["a", "e", "i", "o", "u"])  # Use set for faster lookup
        max_length = 0
        n = len(s)
        
        # Use two pointers to iterate over the string
        left = 0
        
        while left < n:
            vowel_count = {v: 0 for v in vowels_List}  # Track vowel counts
            for right in range(left, n):
                if s[right] in vowels_List:
                    vowel_count[s[right]] += 1
                
                # Check if all vowel counts are even and left, right chars are not vowels
                if all(count % 2 == 0 for count in vowel_count.values()) :
                #    s[left] not in vowels_List and \
                #    s[right] not in vowels_List:
                    max_length = max(max_length, right - left + 1)
            
            left += 1 
        
        return max_length
    

# Testing the Solution class
Test = Solution()
print(Test.findTheLongestSubstring("eleetminicoworoep"))  # Expected output: 13
print(Test.findTheLongestSubstring("leetcodeisgreat"))    # Expected output: 6
print(Test.findTheLongestSubstring("bcbcbc"))             # Expected output: 6
print(Test.findTheLongestSubstring("aeiou"))   
```

```python
# Uncommon Words from Two Sentences

class Solution(object):
    def uncommonFromSentences(self, s1, s2):
        """
        :type s1: str
        :type s2: str
        :rtype: List[str]
        """
        
        s1 = s1.split()
        s2 = s2.split()

        Combine = s1 + s2
        Combine_count =dict(map(lambda x : (x, Combine.count(x)), set(Combine)))

        return [k for k, v in Combine_count.items() if v == 1]
    
Test = Solution()
Test.uncommonFromSentences("this apple is sweet", "this apple is sour")
# Test.uncommonFromSentences("apple apple", "banana")
```

```python
# Longest Substring Without Repeating Characters


# Approach 1 - Set

# We use a set (charSet) to keep track of unique characters in the current substring.
# We maintain two pointers, left and right, to represent the boundaries of the current substring.
# The maxLength variable keeps track of the length of the longest substring encountered so far.
# We iterate through the string using the right pointer.
# If the current character is not in the set (charSet), it means we have a new unique character.
# We insert the character into the set and update the maxLength if necessary.
# If the character is already present in the set, it indicates a repeating character within the current substring.
# In this case, we move the left pointer forward, removing characters from the set until the repeating character is no longer present.
# We insert the current character into the set and continue the iteration.
# Finally, we return the maxLength as the length of the longest substring without repeating characters.

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        charSet = set()
        left = 0
        maxLength = 0
        
        for right in range(len(s)):
            while s[right] in charSet:
                charSet.remove(s[left])
                left += 1
            
            charSet.add(s[right])
            maxLength = max(maxLength, right - left + 1)
        
        return maxLength
    
Test = Solution()

print(Test.lengthOfLongestSubstring("abcabcbb"))  # Expected output: 3
print(Test.lengthOfLongestSubstring("bbbbb"))     # Expected output: 1
        
        
```

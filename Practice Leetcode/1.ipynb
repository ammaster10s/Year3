```python
# Find the Longest Substring Containing Vowels in Even Counts
class Solution(object):
    def findTheLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        # Mapping vowels to bits
        vowels_map = {'a': 0, 'e': 1, 'i': 2, 'o': 3, 'u': 4}
        bitmask = 0
        max_length = 0
        
        # Dictionary to store the first occurrence of each bitmask
        seen = {0: -1}  # We have seen the bitmask 0 at index -1
        
        for i, char in enumerate(s):
            if char in vowels_map:
                # Toggle the corresponding bit for the vowel
                bitmask ^= (1 << vowels_map[char])
            
            if bitmask in seen:
                # If we've seen this bitmask before, calculate the length of the substring
                max_length = max(max_length, i - seen[bitmask])
            else:
                # Otherwise, store the first occurrence of this bitmask
                seen[bitmask] = i
        
        return max_length
    
# findTheLongestSubstring
class Solution(object):
    def findTheLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        vowels_List = set(["a", "e", "i", "o", "u"])  # Use set for faster lookup
        max_length = 0
        n = len(s)
        
        # Use two pointers to iterate over the string
        left = 0
        
        while left < n:
            vowel_count = {v: 0 for v in vowels_List}  # Track vowel counts
            for right in range(left, n):
                if s[right] in vowels_List:
                    vowel_count[s[right]] += 1
                
                # Check if all vowel counts are even and left, right chars are not vowels
                if all(count % 2 == 0 for count in vowel_count.values()) :
                #    s[left] not in vowels_List and \
                #    s[right] not in vowels_List:
                    max_length = max(max_length, right - left + 1)
            
            left += 1 
        
        return max_length
    

# Testing the Solution class
Test = Solution()
print(Test.findTheLongestSubstring("eleetminicoworoep"))  # Expected output: 13
print(Test.findTheLongestSubstring("leetcodeisgreat"))    # Expected output: 6
print(Test.findTheLongestSubstring("bcbcbc"))             # Expected output: 6
print(Test.findTheLongestSubstring("aeiou"))   
```

```python
# Minimum Time Difference
class Solution(object):
    def findMinDifference(self, timePoints):
        """
        :type timePoints: List[str]
        :rtype: int
        """
        # Step 1: Convert time points to minutes
        minutes = []
        for i in range(len(timePoints)):
            minutes.append(int(timePoints[i][:2]) * 60 + int(timePoints[i][3:]))
        
        # Step 2: Sort the list of minutes just once
        minutes.sort()
        
        # Step 3: Initialize the minimum difference to a large number
        min_diff = float('inf')
        
        # Step 4: Compare consecutive times and update the minimum difference
        for i in range(1, len(minutes)):
            min_diff = min(min_diff, minutes[i] - minutes[i-1])
        
        # Step 5: Check the cyclic difference (wrap around the day)
        cyclic_diff = 1440 - minutes[-1] + minutes[0]
        min_diff = min(min_diff, cyclic_diff)
        
        # Step 6: Return the minimum difference
        return min_diff
        
            
            
Test = Solution()
print(Test.findMinDifference(["23:59","00:00"]))  # Expected output: 1
            
```

```python
# Lexicographical Numbers 
class Solution(object):
    def lexicalOrder(self, n):
        num = 1
        ans = []
        while len(ans) < n:
            ans.append(num)

            if num * 10 <= n: num *= 10
            elif num % 10 != 9 and num + 1 <= n: num += 1
            else:
                while (num / 10) % 10 == 9:
                    num /= 10
                num = (num / 10) + 1
        return ans[2]
    
# Start with the number 1.
# Recursively build the next lexicographically smaller number by attempting to multiply the current number by 10. This takes us to the next "level" of lexicographical order.
# If multiplying by 10 overshoots n, we backtrack and try incrementing the current number (i.e., move to the next sibling in the lexicographical tree).
# We repeat this process for numbers from 1 to 9 to cover all lexicographical paths.

Test  = Solution()
print(Test.lexicalOrder(1,1))  # Expected output: [1, 10, 11, 12, 13, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
# Find Kth Smallest Lexicographical Number
class Solution(object):
    def findKthNumber(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: int
        """
        current = 1
        k -= 1  # Convert k to zero-indexed for easier calculation
        
        while k > 0:
            count = self.countNumbers(current, n)
            if k >= count:
                current += 1
                k -= count
            else:
                current *= 10
                k -= 1
        
        return current
    
    def countNumbers(self, prefix, n):
        count = 0
        curr = prefix
        next_curr = curr + 1
        
        while curr <= n:
            count += min(n + 1, next_curr) - curr
            curr *= 10
            next_curr *= 10
        
        return count
```

```python
class Solution(object):
    def waysToSplitArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        answer = 0
        total_sum = sum(nums)
        prefix_sum = 0

        for i in range(len(nums) - 1):
            prefix_sum += nums[i]
            suffix_sum = total_sum - prefix_sum
            if prefix_sum >= suffix_sum:
                answer += 1

        return answer
                      
Test = Solution()
Test.waysToSplitArray([1,1,1,2,3])    
```

```python
from string import ascii_lowercase

class Solution(object):
    def shiftingLetters(self, s, shifts):
        """
        :type s: str
        :type shifts: List[List[int]]
        :rtype: str
        """
        
        letters = list(ascii_lowercase)
        
        n = len(s)
        shift_array = [0] * (n + 1)

        # Apply shifts
        for start, end, direction in shifts:
            if direction == 0:
                shift_array[start] -= 1
                shift_array[end + 1] += 1
            else:
                shift_array[start] += 1
                shift_array[end + 1] -= 1
        
        
        net_shift = 0
        for i in range(n):
            net_shift += shift_array[i]
            shift_array[i] = net_shift
        
        s  =list(s)
        for j in range(n):
            
            newletters = letters.index(s[j])
            s[j] = letters[(newletters + shift_array[j])%26]

        return ''.join(s)
    
Test  = Solution()
Test.shiftingLetters("abxka", [[2,3,1],[0,2,0],[1,4,0]])
Test.shiftingLetters("abc", [[0,1,0],[1,2,1],[0,2,1]])
                
```

```python
class Solution(object):
    
    
    def stringMatching(self, words):
        """
        :type words: List[str]
        :rtype: List[str]
        """
        def constructLps(pat, lps):

            # len stores the length of longest prefix which 
            # is also a suffix for the previous index
            len_ = 0
            m = len(pat)
            
            # lps[0] is always 0
            lps[0] = 0

            i = 1
            while i < m:
                
                # If characters match, increment the size of lps
                if pat[i] == pat[len_]:
                    len_ += 1
                    lps[i] = len_
                    i += 1
                
                # If there is a mismatch
                else:
                    if len_ != 0:
                        
                        # Update len to the previous lps value 
                        # to avoid redundant comparisons
                        len_ = lps[len_ - 1]
                    else:
                        
                        # If no matching prefix found, set lps[i] to 0
                        lps[i] = 0
                        i += 1
        
        
        def search(pat, txt):
            n = len(txt)
            m = len(pat)

            lps = [0] * m
            res = []

            constructLps(pat, lps)

            # Pointers i and j, for traversing 
            # the text and pattern
            i = 0
            j = 0

            while i < n:
                
                # If characters match, move both pointers forward
                if txt[i] == pat[j]:
                    i += 1
                    j += 1

                    # If the entire pattern is matched 
                    # store the start index in result
                    if j == m:
                        res.append(txt)
                        
                        # Use LPS of previous index to 
                        # skip unnecessary comparisons
                        break
                
                # If there is a mismatch
                else:
                    
                    # Use lps value of previous index
                    # to avoid redundant comparisons
                    if j != 0:
                        j = lps[j - 1]
                    else:
                        i += 1
            return res 
    
        result = []
        for i in words:
            for j in words:
                if i != j and i in j:
                    result.append(i)
        
        return list(set(result))
    



Test = Solution()
Test.stringMatching(["mass","as","hero","superhero"])
```

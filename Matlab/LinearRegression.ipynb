```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([0.5,0.8,1.5,2.5,4])  # Independent variable (input values)
y_actual = np.array([1.1,2.4,5.3,7.6,8.9])  # Dependent variable (actual values)

y_pred = 0.2024 * x + 0.0993

y_mean = np.mean(y_actual)

Eq = 10.0604*x**2/(2.0383 + x**2)

SSresidual = np.sum((y_actual - Eq) ** 2)


SStotal = np.sum((y_actual - y_mean) ** 2)

print(f"Residual Sum of Squares (SSresidual): {SSresidual:.4f}")
print(f"Total Sum of Squares (SStotal): {SStotal:.4f}")

plt.scatter(x, y_actual, color="blue", label="Actual Data")
plt.plot(x, y_pred, color="red", label="Predicted Line")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()

print(f"R-squared: {1 - (SSresidual / SStotal):.4f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
c = np.array([0.5, 0.8, 1.5, 2.5, 4.0])  # Oxygen concentration (mg/L)
k = np.array([1.1, 2.4, 5.3, 7.6, 8.9])  # Growth rate (per time unit)

# Transform the data for linearization
x = 1 / c**2  # Independent variable for linear regression
y = 1 / k     # Dependent variable for linear regression

# Reshape x for linear regression
x = x.reshape(-1, 1)

# Perform linear regression
model = LinearRegression()
model.fit(x, y)

# Get the slope (m) and intercept (b) from the model
m = model.coef_[0]
b = model.intercept_

# Estimate c_s and k_max
k_max = 1 / b
c_s = m * k_max

print(f"Estimated c_s: {c_s:.4f}")
print(f"Estimated k_max: {k_max:.4f}")

# Calculate the R^2 value
r_squared = model.score(x, y)
print(f"R^2: {r_squared:.4f}")

# Predict the growth rate at c = 2 mg/L
c_predict = 2
k_predict = (k_max * c_predict**2) / (c_s + c_predict**2)
print(f"Predicted growth rate at c = 2 mg/L: {k_predict:.4f}")

# Optional: Plot the data and the fitted line
plt.scatter(x, y, color="blue", label="Data")
plt.plot(x, model.predict(x), color="red", label="Fitted Line")
plt.xlabel("1/c^2")
plt.ylabel("1/k")
plt.legend()
plt.show()
```
